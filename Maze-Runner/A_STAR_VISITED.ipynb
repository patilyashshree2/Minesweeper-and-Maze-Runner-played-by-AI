```python
import random
def genMaze(dim,p):
    maze = []
    for x in range(dim):
        row = []
        for y in range(dim):
            if random.random() > p:
                row.append(0)
            else :
                row.append(1)
        maze.append(row)
    maze[0][0] = 0
    maze[dim-1][dim-1] = 0
    return maze
```

```python
maze = genMaze(10,0.20)
```

```python
def draw(maze):
  blocked = "⬛"
  free = "⬜"
  for row in maze:
    tiles = ""
    for tile in row:
      tiles += blocked if tile==1 else free
    print(tiles)
```

```python
draw(maze)
```
